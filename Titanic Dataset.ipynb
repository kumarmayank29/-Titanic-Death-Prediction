# Data Cleaning

## Removing the NULL values

### We want to fill missing age values instead of dropping the entire age column.One way to fill missing values is by filling the mean age of all pasengers(imputation).However we could be smarted by this by chceking the  average age based on pasenger class .For this we use box plot as 50th percentile of Boxplot gives us avg age 

```python
train.columns
```

```python
sns.boxplot(x='Pclass',y='Age',data=train)
# We can clearly observe passenger of class 1 have avg age of - 37
#                        passenger of class 2 have avg age of - 29
#                        passenger of class 3 have avg age of - 24

#So replace NULL values of age based on Pasenger class with the above mean
```

```python
type(train['Age'])
```

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if(pd.isnull(Age)):
        
        if Pclass == 1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age;
        
```

```python
# NOW APPLY THE ABOVE DEFINED FUNCTION
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1)
```

Now check the Heatmap again to see all nul, values of age got vanished

```python
sns.heatmap(train.isnull(),yticklabels=False) #NuLL values of age are now vanished
```

```python
#There are many many NULL values so to replace NAN  values in cabin Alot of feature engineering need to be applied we will 
# do that later for now just drop that column
train.drop('Cabin',axis=1,inplace=True)
```

### We need to convert categorical features to dummy variable using pandas otherwise our machine learning algorithm will not be able to directly take in those features as inputs

```python
pd.get_dummies(train['Embarked']).head(10)
#Numbers are assigned as many categories are there.Suppose here we have three categories  C, Q,S whichever holds true is given
# 1 others are given 0.Note No of Dummy variables formed equal to toal number of categories 
#However duringan;laysis always one dummy variable is dropped .For Example if we have 100 dummy variable we will consider only 99
# SEE CONCEPT OF DUMMY VARIABLE KERILL ERMENKO
```

```python
embark=pd.get_dummies(train['Embarked'],drop_first=True)

#We remove first column because other two column can represnt first column.This is because see above values of binary numbers
# we can clearly see first column value can have code as for 


# For S we have 0 1
# For Q we have 1 0
# For C we have 0 0
#This is called DUMMY VARIABLE TRAP CONCEPT
```

## Building a Logistic Regression model

## Train Test Split


```python
train_data=train.drop('Survived',axis=1)
```

```python
test_data=train['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(train_data,test_data,test_size=0.2,random_state=10)
```

## Training and Prediction

```python
from sklearn.linear_model import LogisticRegression
Regressor=LogisticRegression()

```

```python
Regressor.fit(X_train,y_train)
```

```python
y_pred=Regressor.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix

accuracy=confusion_matrix(y_test,y_pred)
print(accuracy)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```
